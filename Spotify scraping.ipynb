Seems like we need to delete any occurrence of [Featuring, & and apostrophes] in the artist names and song titles in order to get results from the API.

For the songs containign a slash, we'll only take the part before that.

For the artist field containing ' x ' (which means Featuring), we remove the ' x '.

```python
missing['Performer_x'] = missing['Performer_x'].str.replace('&','')
missing['Performer_x'] = missing['Performer_x'].str.replace('\'','')
missing['Song_x'] = missing['Song_x'].str.replace('\'','')
missing['Performer_x'] = missing['Performer_x'].str.replace('Featuring','')

missing['Song_x'] = missing['Song_x'].str.split('/').str[0]
missing['Song_x'] = missing['Song_x'].str.replace(r"\(.*?\)","")

missing['Performer_x'] = missing['Performer_x'].str.replace(' x ',' ')
missing['Performer_x'] = missing['Performer_x'].str.replace(' X ',' ')
```

```python
for song in missing.values:
    artist = song[4]
    song = song[3]
    result = sp.search(q=f'artist:{artist} track:{song}')
    if len(result['tracks']['items']) < 1:
        print(artist, song)
```

```python
def fill_id(track):
    if(track.Performer_x == 'Duane Eddy His Twangy Guitar And The Rebels'):
        return '4nAJtcmiyoL0ARY5WZY9IN'
    else: 
        result = sp.search(q=f'artist:{track.Performer_x} track:{track.Song_x}')
        if len(result['tracks']['items']) > 0:
            print(result['tracks']['items'][0]['id'])
            return result['tracks']['items'][0]['id']
        else: return

missing['spotify_track_id'] = missing.apply(lambda track:fill_id(track),axis=1)
```

```python
missing.isnull().sum()
```

```python
# We will filter out the 98 missing tracks
missing = missing[~missing['spotify_track_id'].isnull()].reset_index(drop=True)
```

# Get audio features for missing songs

sp.audio_features does not accept more than 100 ids, so we need to do 100 by 100

```python
i = 0
audio_features = pd.io.json.json_normalize(sp.audio_features(missing['spotify_track_id'].values[i:i+100]))
while (i+100 < len(missing)):
    i = i + 100
    audio_features = audio_features.append(pd.io.json.json_normalize(sp.audio_features(missing['spotify_track_id'].values[i:i+100])), ignore_index=True)

plus = len(missing) % 100
audio_features = audio_features.append(pd.io.json.json_normalize(sp.audio_features(missing['spotify_track_id'].values[len(missing)-plus:len(missing)])), ignore_index=True)
```

```python
audio_features = audio_features.drop(columns=['analysis_url', 'track_href', 'uri', 'type'])
```

```python
cols_to_use = audio_features.columns.difference(missing.columns)
complete = missing.drop(columns='_merge').merge(right=audio_features, how='left', left_on='spotify_track_id', right_on='id', indicator=True, suffixes=('_y', '')).drop_duplicates().reset_index(drop=True)
complete = complete.drop(complete.filter(regex='_y$').columns.tolist(),axis=1)
complete
complete.to_csv('scraped_100.csv')
```
