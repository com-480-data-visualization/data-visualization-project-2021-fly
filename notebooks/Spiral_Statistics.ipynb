```python
import pandas as pd
```

```python
complete_100 = pd.read_csv('./data/complete_100.csv.zip', compression='zip')
```

```python
complete_100.spotify_genre = complete_100.spotify_genre.apply(lambda x: (str(x).strip('][').split(', ')))
```

```python
complete_100['spotify_track_album'] = complete_100['spotify_track_album'].astype(str)
```

```python
complete_100 = complete_100.drop(columns=['spotify_genre']).drop_duplicates()
```

#### Top Albums

```python
# ALl rankings
complete_100.groupby(by=['spotify_track_album', 'Performer_x']).count().sort_values(by='url', ascending=False).url.head(15)
```

```python
# Most weeks at number 1
complete_100[complete_100['Week Position'] == 1].groupby(by=['spotify_track_album']).count().sort_values(by='url', ascending=False).head(15)
```

#### Top Songs

```python
# Most weeks on Billboard
# ALl rankings
complete_100.groupby(by=['SongID']).count().sort_values(by='url', ascending=False).url.head(15)
```

```python
# Most weeks at number 1
complete_100[complete_100['Week Position'] == 1].groupby(by=['SongID']).count().sort_values(by='url', ascending=False).head(15)
```

#### Top Artists

```python
# ALl rankings
complete_100.groupby(by=['Performer_x']).count().sort_values(by='url', ascending=False).url.head(15)
```

```python
# Most weeks at number 1
complete_100[complete_100['Week Position'] == 1].groupby(by=['Performer_x']).count().sort_values(by='url', ascending=False).head(15)
```
